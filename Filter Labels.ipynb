```python
data_path='/home/ambika/INF_project/filtered_one_level_labels_all.txt'
data = np.loadtxt(data_path, str, delimiter='\t')

filename='filtered_labels_all.txt'
b = open(filename, 'wb') 
a = csv.writer(b)

for i in range(0,99):
    d=list(data[i][1:])
    #print d
    t=[data[i][1],data[i][3],data[i][5],data[i][7],data[i][9]]
    #print t
    t2=list(set(t))
    #print t2
    if '' in t2:
        t2.remove('')
    #print t2
    final=[]
    for u in range(0,len(t2)):
        ind=[]
        for v in range(0,len(d)):
            if t2[u]==d[v]:
                ind.append(v)
                #print ind
            prob=0
            for w in range(0,len(ind)):
                prob=prob+float(d[ind[w]+1])
        #print prob
        final.append(t2[u])
        final.append(str(prob))
    
    print final
    var2=data[i][0]
    for j in range(len(final)):
        var2+='\t'+final[j]
    print var2
    a.writerows([[var2]])

b.close()
```
