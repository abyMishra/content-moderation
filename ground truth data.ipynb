```python
%matplotlib inline
from pycocotools.coco import COCO
import numpy as np
import csv
import skimage.io as io
import matplotlib.pyplot as plt
import pylab
import urllib
pylab.rcParams['figure.figsize'] = (10.0, 8.0)

dataDir='/home/ambika/INF_project/'
dataType='val2014'
annFile='%s/annotations/instances_%s.json'%(dataDir,dataType)

# initialize COCO api for instance annotations
coco=COCO(annFile)

final_path='/home/ambika/INF_project/final_image_paths.txt'
image_list = np.loadtxt(final_path, str)

filename='ground_truth.txt'
b = open(filename, 'wb') 
a = csv.writer(b)

for p in range(0,1000):
    x_init=''
    for q in range(0,len(image_list[0])):
        #print q
        #print lines[1][q]
        if image_list[p][q]=='.':
            break
        elif (image_list[p][q]=='1' or image_list[p][q]=='2' or image_list[p][q]=='3' or image_list[p][q]=='4' or image_list[p][q]=='5' or image_list[p][q]=='6' or image_list[p][q]=='7' or image_list[p][q]=='8' or image_list[p][q]=='9' or image_list[p][q]=='0'):
            x_init+=image_list[p][q]
    x_init=x_init
    print x_init
    imgId=int(x_init);
    annIds = coco.getAnnIds(imgIds=imgId)
    l=len(annIds)
    anns = coco.loadAnns(annIds)
    #coco.showAnns(anns)
    #print annIds
    #print anns
    #print anns[0]["category_id"]
    var=x_init+'\t'
    for i in range(0,l):
        var+=str(anns[i]["category_id"])+'\t'+str(anns[i]["bbox"])+'\t'
    print var
    a.writerows([[var]])

    
b.close()
```
